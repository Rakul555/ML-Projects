```python
data=pd.read_csv('Dataset .csv')
data.head()
```

```python
data['City'] = data['City'].replace('Bras�_lia', 'Brasília')
data['City'] = data['City'].replace('S��o Paulo','São Paulo')
data['City'] = data['City'].replace('��stanbul','Istanbul')

```

```python
data.loc[23,"Restaurant Name"]
```

```python
data['City'] = data['City'].apply(lambda x: unicodedata.normalize('NFKD', x).encode('ascii', 'ignore').decode('ascii'))
data['Address'] = data['Address'].apply(lambda x: unicodedata.normalize('NFKD', x).encode('ascii', 'ignore').decode('ascii'))
data['Locality'] = data['Locality'].apply(lambda x: unicodedata.normalize('NFKD', x).encode('ascii', 'ignore').decode('ascii'))
data['Locality Verbose'] = data['Locality Verbose'].apply(lambda x: unicodedata.normalize('NFKD', x).encode('ascii', 'ignore').decode('ascii'))

```

**Task 1**

```python
from sklearn.preprocessing import LabelEncoder

lable=LabelEncoder()
data['City']=lable.fit_transform(data['City'])
```

```python
data['Locality']=lable.fit_transform(data['Locality'])
data['Rating text']=lable.fit_transform(data['Rating text'])
data['Cuisines']=lable.fit_transform(data['Cuisines'])
data['Currency']=lable.fit_transform(data['Currency'])
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(data[['City','Locality','Country Code','Cuisines','Price range','Votes']],data['Aggregate rating'],test_size=0.2)
```

```python
from sklearn.tree import DecisionTreeRegressor
model=DecisionTreeRegressor()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
from sklearn.metrics import mean_squared_error
y_pred=model.predict(X_test)
msr=mean_squared_error(y_test,y_pred)
msr
```
